```python
from pandas.core.groupby.ops import base
dataset = []
for page in range(1,90):
  url = f"https://www.flipkart.com/redmi-note-10-pro-vintage-bronze-128-gb/product-reviews/itmb4b28cb04dce7?pid=MOBGB725PZQYDN4D&lid=LSTMOBGB725PZQYDN4D6XA65V&marketplace=FLIPKART&page={page}"
  browser.get(url)
  xx = browser.find_elements(By.XPATH,"//span[@class='_1BWGvX']")
  d_x = [click(x) for x in xx]
  containers = [x.get_attribute('innerHTML') for x in browser.find_elements(By.XPATH,"//div[@class='col _2wzgFH K0kLPL']")]

  #print(len(containers))
  for container in containers:
    bsObj = BeautifulSoup(container)
    rev = bsObj.find("div",attrs={"class":"t-ZTKy"}).text.strip()
    rev = rev[:-9]
    

    dt = bsObj.find("div",attrs={"class":"row _3n8db9"}).text.strip()
    
    try:
      try:
        rating = bsObj.find("div",attrs={"class":"_3LWZlK _1BLPMq"}).text.strip() #3,4,5 star ratings
        rating = float(rating)
      except:
        rating = bsObj.find("div",attrs={"class":"_3LWZlK _1rdVr6 _1BLPMq"}).text.strip() #1 star rating
        rating = float(rating)
    except:
      rating = bsObj.find("div",attrs={"class":"_3LWZlK _32lA32 _1BLPMq"}).text.strip() #2 star ratings
      rating = float(rating)
    

    dataset.append((dt,rev,rating))
```

```python
data_flipkart = pd.DataFrame(dataset,columns = ["Dates","Reviews","Ratings"])
```
